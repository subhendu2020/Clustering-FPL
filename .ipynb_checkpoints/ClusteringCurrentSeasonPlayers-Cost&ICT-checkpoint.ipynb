## Clustering With ICT & Cost

```python
dummydata=pd.get_dummies(dfCurrentPlayer['Position'],prefix='Postion')
dfCurrPlayercat=pd.concat([dfCurrentPlayer,dummydata],axis=1)
dfCurrPlayercat.drop('Position',axis=1,inplace=True)

# ClusterCols=['Cost','ICT','Postion_DEF','Postion_FWD','Postion_GK','Postion_MID']
# Doing without Position
ClusterCols=['Cost','ICT']

```

```python
from sklearn.preprocessing import QuantileTransformer

# Quantile Transformer balances out the presence of Outliers. Hence used. Moreover it brings the numerical values between 0-1.
# In this case there are also categorical variables, converted to 0-1
```

```python
scaleqts=QuantileTransformer()
scaleqts.fit(dfCurrPlayercat[ClusterCols])
scaledata=scaleqts.transform(dfCurrPlayercat[ClusterCols])
```

```python
dfTransformedCluster=pd.DataFrame(data=scaledata,columns=ClusterCols)
# dfTransformedCluster.head()
```

```python
# Use of Dendrogram for no. of clusters
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(15,10))
clusterK=shc.dendrogram(shc.linkage(dfTransformedCluster,method='ward'))
plt.axhline(y=4.5,color='r',linestyle='--')
```

### Identifying K i.e. No of Clusters

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
```

```python
predictedCluster=cluster.fit_predict(dfTransformedCluster)
```

```python
dfCurrentPlayer['ClusterLabel']=pd.Series(predictedCluster,index=dfCurrentPlayer.index)
dfCurrentPlayer.head()
```

### Cluster Concentration

```python
dfCurrentPlayer.groupby('ClusterLabel')['ID'].count()
```

### Cluster Definition

```python
dfCurrentPlayer.groupby('ClusterLabel')[clusterNumCols].agg(["mean",'std']).reset_index()
```

```python
dfCurrentPlayer.groupby(['ClusterLabel','Position'])['ID'].count()
```

```python
plt.figure(figsize=(15,10))
sn.scatterplot(x='Cost',y='ICT',hue='ClusterLabel',data=dfCurrentPlayer,s=200)
```
